```python
#Importamos las librerias pandas y numpy para trabajr con los datos
import numpy as np
import pandas as pd
#Importamos las librerias matplotlib y seaborn para trabajar los graficos
import matplotlib.pyplot as plt
import seaborn as sns
#Desde la libreria de archivos de google.colab subiremos el archvios csv de los datos
from google.colab import files
files.upload()
```

```python
#Leemos el archivos csv subiod y lo guardaos en una variable data_autos donde se guarda el dataframe
data_autos=pd.read_csv("/content/autos - autos.csv", delimiter=";")
#Mostramos las ultimas 100 filas del data frame
data_autos.tail(100)
```

```python
#Definimos la funcion change_value con dos parametro uno que es el dataframe (df) y otro el nombre de la columna a cambiar sus datos ordinales a cardinales
def change_value(df,column):
  #Se define una variable dfarray donde se guardaran los datos de los datos que no se encuentran repetidos de la variable categorica
  dfarray=pd.DataFrame(df[column].unique())
  #Se define una funcion denominada indice donde pasa de parametro el dato a cambiar
  def indice(x):
    #Se realiza un bucle desde 0 hasta la longitud del aray
    for i in range(len(dfarray)):
      #Definimos la condicional , si el dato dato se encuentra dentro del recorrido del dataframe dfarray que imprima el indice +1
      if(x==dfarray[0][i]):
        return i+1
  #Se pasa todos los valores de la columna elegida en el dataframe y se le aplica la propiedad appy teniendo como argumento la funicion indice creada anteriormente      
  df[column]=df[column].apply(indice)
  #Retorna el nuevo dataframe con los cambios realizados
  return df
change_value(data_autos,'Pais_Origen')
```

```python
#Realizamos la  limpieza de datosbuscando datos nulos dentro del dataframe
data_autos.isnull().sum()
#Mostramos la informacion del dataframe ya realizado con la limpieza de los datos nulos, observando que se cumplen las caracteristicas solicitadas de la gestion de datos
data_autos.info()
#Se elimina los datos nulos del dataframe y se guarda en la misma variable creada para el dataframe
data_autos=data_autos.dropna()
#Se crea el campo var_1 en el cual no la libreria numpy realizaremos las aoperaciones solicitadas para las primeras 5 variables, en la primera se utiliza np.log10
data_autos['var_1']= np.log10(data_autos['Millas_Por_Galon']/data_autos['Cilindros'])
#Se crea var_2 en donde se utiliza np.sqrt y np.exp para hallar la raiz cuadrada y elevar e al numero solicitado y calcular la varibale solicitada
data_autos['var_2']= np.sqrt(data_autos['Millas_Por_Galon'])*np.exp(data_autos['Cilindros'])/200
#Se crea var_3 en donde se utiliza np.select el cual selecciona un cierto valor a partir de condicionales establecidas (Cabe indicar que los valores en este campo son los mismos es por ello cuando se realizo el mapa de calor no se toma en cuenta)
data_autos['var_3']=np.select([data_autos['Millas_Por_Galon']>data_autos['Cilindros'],data_autos['Millas_Por_Galon']<data_autos['Cilindros']],[5,3],3)
#Se crea var_4 en donde se utiliza np.log que es el calculo de loagritmo neperiano
data_autos['var_4']=np.log(1/(data_autos['Millas_Por_Galon']/data_autos['Cilindros'])*100)
#Se crea var_5 en donde se utiliza np.divide 
data_autos['var_5']=np.divide(data_autos['Cilindros']**2,data_autos['Millas_Por_Galon'])
data_autos
```

```python
#Se define la funcion graficar_dataframe y se pasan como parametro el dataframe(df), la variable cuantitativa (columna con datos numericos), la variable cualitativa (columna con datos categoricos)
def graficar_dataframe(df, var_cuantitativa, var_cualitativa):
  #Para graficar el de cajas y bigotes seleccionamos el marco y el tamaño de la grafica
  fig, ax=plt.subplots(figsize=(10,6))
  #Llamamos a seaborn con el metodo boxplot, e indicamos los valores correspondientes para su grafica
  sns.boxplot(x=var_cualitativa, y=var_cuantitativa, data=df, ax=ax)
  #Seteamos el titulo de la grafica
  ax.set_title('Graficos de cajas y bigotes de ' + var_cuantitativa + ' por ' + var_cualitativa)
  plt.savefig("BoxPlot_"+var_cuantitativa + var_cualitativa+".jpg")
  #Para graficar barras de la media en el marco y el tamaño de la grafica
  fig1, ax1=plt.subplots(figsize=(10,6))
  #Llamamos a seaborn con el metodo barplot, e indicamos los valores correspondientes para su grafica, como es de la media colocamos el estimator=np.mean
  sns.barplot(x=var_cualitativa, y=var_cuantitativa, estimator=np.mean, data=df, ax=ax1)
  #Seteamos el titulo de la grafica
  ax1.set_title('Graficos de barras de la media de ' + var_cuantitativa + ' por ' + var_cualitativa)
  plt.savefig("BarPlot_"+var_cuantitativa + var_cualitativa+".jpg")
  #Llamamos a seaborn con el metodos displot para grafica el histograma y con kde en True visualizamos la curva de cada histograma
  sns.displot(data=df, x=var_cuantitativa, hue=var_cualitativa, kde=True)
  #Seteamos el titulo de la grafica
  plt.title('Histograma con curvas de ' + var_cuantitativa + ' por ' + var_cualitativa)
  plt.savefig("DisPlot_"+var_cuantitativa + var_cualitativa+".jpg")
#Para el ejemplo utilizamos la variables Millas_Por_Galon y Pais_Origen, los datos de este ultimo se muestran en numero debido al cambio de valor realizado en el ejercicio 1
graficar_dataframe(data_autos,'Desplazamiento','var_6')
```

```python
#Creamos los dataframe para cada uno de los paises a ralizar el mapa de calor
datapais1=data_autos[data_autos['Pais_Origen']==1].drop(columns=['Pais_Origen'])
datapais2=data_autos[data_autos['Pais_Origen']==2].drop(columns=['Pais_Origen'])
datapais3=data_autos[data_autos['Pais_Origen']==3].drop(columns=['Pais_Origen'])
#Definimos el area del grafico
fig1, ax1 = plt.subplots(1,1, figsize=(15,15)) # con figsize es posible configurar el tamaño de la imagen
#Llamamos a seaborn para ejecutar el mapa de calor y con el datafram de cada pais realizamos la correlacion y redondeamos a 2 , el estilo van a realizar en tonos azules
sns.heatmap(
    datapais1.corr().round(2), # redondeo a 1 decimal para una mejor visualización
    cmap="Blues",
    annot=True, 
    ax=ax1)
ax1.set_title('Mapa de Calor de Dataframe para USA')
plt.savefig("HeadMap_data_autos_USA.jpg")
fig2, ax2 = plt.subplots(1,1, figsize=(15,15)) # con figsize es posible configurar el tamaño de la imagen
sns.heatmap(
    datapais2.corr().round(2), # redondeo a 1 decimal para una mejor visualización
    cmap="Blues",
    annot=True, 
    ax=ax2)
ax2.set_title('Mapa de Calor de Dataframe para EUROPE')
plt.savefig("HeadMap_data_autos_EUROPE.jpg")
fig3, ax3 = plt.subplots(1,1, figsize=(15,15)) # con figsize es posible configurar el tamaño de la imagen
sns.heatmap(
    datapais3.corr().round(2), # redondeo a 1 decimal para una mejor visualización
    cmap="Blues",
    annot=True, 
    ax=ax3)
ax3.set_title('Mapa de Calor de Dataframe para JAPAN')
plt.savefig("HeadMap_data_autos_JAPAN.jpg")

```
