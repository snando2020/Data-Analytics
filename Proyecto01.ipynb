**Ejercicio N°1:**

Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.

```python
# Importamos la librería pandas para trabajar con dataframe
import pandas as pd
# Importamos la libreris date y timedelta para trabajar con fechas
from datetime import date , timedelta
#creamos una variable ventas donde se almacenara el dataframe con las columnas requeridas
ventas=pd.DataFrame(columns=['fecha', 'producto', 'cantidad','precio_unitario','total','descuento', 'total_con_descuento'])
#Realizamo un bucle con un rango de 5 días
for i in range(5):
  # La variable variacion especifica los dias a restar a la fecha actual
  variacion=timedelta(i)
  # Variable producto donde se introduce el producto por dia
  producto=input('Introduce el producto del día ' + str(date.today()-variacion) +': ')
  # Variable cantidad donde se introduce la cantidad del producto por dia
  cantidad=input('Introduce la cantidad:')
  #Variabe punitario donde se introduce el precio unitario por dia
  punitario=input('Introduce el precio unitario:')
  #Varibale fecha donde se determina las fechas a ingresar desde el dia actual
  fecha=str(date.today()-variacion)
  #Se ingresa una fila de datos con los campos de las columnas requeridas y de acuerdo a las variables solicitadas por el input
  ventas=ventas.append({'fecha': fecha, 'producto': producto,'cantidad': int(cantidad),'precio_unitario': float(punitario),'total': int(cantidad)*float(punitario),'descuento':'5%','total_con_descuento':int(cantidad)*float(punitario)*0.95}, ignore_index=True)
#Al finalizar el bucle de los 5 dias se establece el indice por la columna fecha
ventas.index=ventas['fecha']
#Se elimina la columna fecha
ventas = ventas.drop(columns=['fecha'])
#Se muestra el dataframe
ventas
```

**Ejercicio Nº 2**

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.

```python
# Importamos la librería pandas para trabajar con dataframe
import pandas as pd
# Definimos las función descriptiva_ingresos el cual genera una serie con el maximo, minimo y media de un dataframe 
def descriptiva_ingresos(ingresosclientes):
    ingresosclientes = pd.Series(ingresosclientes)
    datosestadisticos = pd.Series([ingresosclientes.min(), ingresosclientes.max(), ingresosclientes.mean()], index=['Minimo', 'Maximo', 'Media'])
    return datosestadisticos
# Se define el diccionario con los ingresos mensuales de los clientes
clientes = {'Cliente1':95245.00, 'Cliente2':12548.20, 'Clientes3':25624.12, 'Clientes4': 1128.50, 'Clientes5': 500.00, 'Cliente6':245.30, 'Cliente7':578.40, 'Clientes8':104002.25, 'Clientes9': 0.00, 'Clientes10': 27550.35}
# Imprime los resultados de la función
print(descriptiva_ingresos(clientes))
```

**Ejercicio Nº 3**

Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:

|**Mes**|**Ventas**|**Gastos**|
|-------|----------|----------|
|Enero  |30500     |22000     |
|Febrero|35600	   |23450     |
|Marzo	|28300	   |18100     |
|Abril	|33900     |35700     |
|Mayo   |42500     |32450     |

A continuación, generar nuevas columna:
1.	Columna 1 =  Gasto/Ventas*100.
2.	Columna 2 =  Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3.	Columna 3 =  Si Gasto>Ventas, calcular el sobre gasto.
4.	Obtener el total de las ventas de los últimos 3 meses
5.	Obtener el total de gastos de febrero a abril


```python
# Importamos la librería pandas para trabajar con dataframe
import pandas as pd
# Importamos la librería numpy para trabajar con gran volumenes de datos 
import numpy as np
# Se creo el dataframe con los datos dados
datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril','Mayo'], 'Ventas':[30500, 35600, 28300, 33900, 42500], 'Gastos':[22000, 23450, 18100, 35700, 32450]}
# Ser crea el dataframe ratiosoperativos
ratiosoperativos = pd.DataFrame(datos)
# Se genera la columna 1 gastos/ventas*100
ratiosoperativos['Columna_1']=100*ratiosoperativos['Gastos']/ratiosoperativos['Ventas']
# Se genera la columna 2 a traves de la condicional de que las ventas sea mayor a 30000 e imprima meta superada de lo contario meta no superada
ratiosoperativos['Columna_2'] = ["Meta Superada" if a >=30000 else "Meta no Superada" for a in ratiosoperativos['Ventas']]
# Se genera la columna 3 el cual a traves de la libreria numpy realizar un where con la condicion de que Gastos sea Mayor a Ventas lo cual imprime el sobregasto
ratiosoperativos['Columna_3'] = np.where(ratiosoperativos['Gastos'] > ratiosoperativos['Ventas'], ratiosoperativos['Gastos']-ratiosoperativos['Ventas'], '') 
# Imprime el dataframe con todas las columnas creadas
print(ratiosoperativos)
# Se define la funcion de ventastotal en cual pasamos dos argumentos que son el dataframe y los menes de calculo
def ventastotal(listas, meses):
    return listas.set_index('Mes').loc[meses,'Ventas'].sum()
# Imprime las ventas de los ultimos tres meses de los datos, aca tambine podemos trabajar hallando la fecha maxima y haciendo un bucle con for co un rango de 3 
print('Las ventas de los ultimos 3 meses fue ' + str(ventastotal(ratiosoperativos, ['Marzo','Abril', 'Mayo'])))
# Se define la funcion gastototal de los meses seleccionados
def gastototal(listas, meses):
    return listas.set_index('Mes').loc[meses,'Gastos'].sum()
#Imprime los gastos de los meses seleccionados
print('Los gastos de los meses seleccionados es ' + str(gastototal(ratiosoperativos, ['Febrero', 'Marzo','Abril'])))
```
